## Identifying hurricane season using for loops and if statements

Warm ocean water provides the energy for hurricanes to form. Almost 70 years ago, scientists realized that hurricanes can form only when the surface of the ocean is warmer than 79.7°F, or 26.5°C. Below this temperature, hurricanes will not form or will weaken rapidly.

I've downloaded sea surface temperatures (SSTs) for a location in the tropical Atlantic, northeast of Puerto Rico, where hurricanes tend to form. These are monthly values (January-December) for a typical, or "climatological," year.

Use *for* loop(s) and *if* statement(s) to answer these questions:
* Question 1: What is the average SST over the 12 months? Round the answer to one decimal place using round().
* Question 2: How many months are favorable for hurricane formation?
* Question 3: What month is most favorable for hurricanes to form?
* Question 4: When does hurricane season start and end?


```
# Ocean temperatures at (20°N, 60°W)
#   from NOAA 5km SST climatology:
#   https://coralreefwatch.noaa.gov/product/5km/

months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']
sst_clim = [25.52,25.14,25.06,25.49,26.27,26.96,
            27.48,27.87,28.04,27.97,27.26,26.38]
```

```
# Write your main code here:
sum_sst = 0.0                   # for Question 1
num_hot_months = 0              # for Question 2
current_max_temp = 0.0          # for Question 3
current_max_temp_month = ''     # for Question 3
hurricane_season_months = []    # for Question 4

for mo_idx in range(len(months)):   
  temp = sst_clim[mo_idx]   # retrieving and saving these variables makes our life a bit easier
  month = months[mo_idx]

  # for Question 1:
  sum_sst += temp

  # for Questions 2 and 4:
  if temp > 26.5:   # check if month is warmer than 26.5°C, the hurricane threshold
    num_hot_months += 1                    # for Question 2
    hurricane_season_months.append(month)  # for Question 4

  # for Question 3:
  # update maximum temp if necessary
  if temp > current_max_temp:
    current_max_temp = temp
    current_max_temp_month = month

# for Question 1: calculate average temperature AFTER the loop!
avg_sst = round(sum_sst / 12, 1)   # or sum_sst / len(sst_clim)

# for Question 4: identify first and last hurricane months
hurricane_season_start = hurricane_season_months[0]
hurricane_season_end = hurricane_season_months[-1]

# Print your answers here:
print('Question 1: The average SST is',str(avg_sst) + '°C')
print('Question 2: There are',num_hot_months,'hurricane-favorable months.')
print('Question 3: The warmest, most hurricane-favorable month is ' 
      + current_max_temp_month + '.')
print('Question 4: Hurricane season starts in ' + hurricane_season_start
      + ' and ends in ' + hurricane_season_end + '.')
```

## Hurricane season, except with NumPy instead of loops!

Use NumPy functions to answer these questions:
* Question 1: What is the average SST over the 12 months? Round the answer to one decimal place using round().
* Question 2: How many months are favorable for hurricane formation?
* Question 3: What month is most favorable for hurricanes to form?
* Question 4: When does hurricane season start and end?

```
import numpy as np   # Note that we try to put import statements at the top

# Ocean temperatures at (20°N, 60°W)
#   from NOAA 5km SST climatology:
#   https://coralreefwatch.noaa.gov/product/5km/

months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']
sst_clim = [25.52,25.14,25.06,25.49,26.27,26.96,
            27.48,27.87,28.04,27.97,27.26,26.38]
temp_threshold = 26.5   # units: °C

# First step: convert the lists to arrays
#   (you can use the same variable names, or different ones)
months = np.array(months)
sst_clim = np.array(sst_clim)

# Write your calculations here:

# Question 1
avg_sst = np.mean(sst_clim)

# Question 2
num_hot_months = np.sum(sst_clim > temp_threshold)   # sum of a Boolean array counts the number of True elements
num_hot_months = len(sst_clim[sst_clim >= 26.5])     # length of the list of hurricane-favorable temperatures is the # of hurricane months

# Question 3
max_sst = np.max(sst_clim)

# From the line above, here is option 1:
max_temp_month = months[sst_clim == max_sst]
print(max_temp_month)      # will this work?
print(max_temp_month[0])   # no, we need to index into max_temp_month because it's a list of length 1
max_temp_month = max_temp_month[0]

# From that line, here is option 2:
max_sst_index = np.where(sst_clim == max_sst)
print(max_sst_index)         # note weird tuple format here, which requires indexing
print(max_sst_index[0][0])   # keep indexing using [0] until you get a single number, if you're expecting a single index
max_sst_index = max_sst_index[0][0]
max_temp_month = months[max_sst_index]

# Question 4
fav_months = months[sst_clim > temp_threshold]
hurricane_season_start = fav_months[0]
hurricane_season_end = fav_months[-1]

# Print your answers here:
print('Question 1: The average SST is',str(avg_sst) + '°C')
print('Question 2: There are',num_hot_months,'hurricane-favorable months.')
print('Question 3: The warmest, most hurricane-favorable month is ' 
      + max_temp_month + '.')
print('Question 4: Hurricane season starts in ' + hurricane_season_start
      + ' and ends in ' + hurricane_season_end + '.')
```
