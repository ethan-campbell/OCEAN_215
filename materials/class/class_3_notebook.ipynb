```
rand_string = 'We are learning Python. Python programming today is a wizard at math. Python is now a language that is used everywhere. It is really powerful and fun.'

print(rand_string)
```

```
sentences = rand_string.count('.')
print(sentences)

delimiter = ' '
str_list = rand_string.split(delimiter)
new_list = str_list[0:len(str_list):3]
# new_list = str_list[::3]
print(new_list)

short5_list = new_list[2:7]
# midpt = int(len(new_list)/2)
# shortr_list = new_list[midpt-2:midpt+3]
del new_list[2:7]

print(short5_list)

short5_list.reverse()
print(short5_list)

short5_list.insert(0,'Katy')
print(short5_list)

final_str = delimiter.join(short5_list)
print(final_str)
print(len(final_str))
```
